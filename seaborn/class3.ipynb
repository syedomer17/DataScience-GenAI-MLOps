```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.histplot(tips["total_bill"],bins=30,kde=True,color="skyblue")
plt.title("Histogram of total bills")
```

```python
sns.kdeplot(tips["total_bill"], fill=True,color="green")
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.rugplot(x =  tips["total_bill"] , height = 0.6)
```

```python
sns.histplot(x = "total_bill", data = tips, hue = "sex", kde= True, bins=10, palette="Set2")
```
