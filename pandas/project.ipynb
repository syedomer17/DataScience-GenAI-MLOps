```python
def level(x): 
    if x >= 90: 
        return "High" 
    elif x >= 70: 
        return "Medium" 
    elif x < 70: 
        return "Low"
```

```python
df["scoring_level"] = df["total_score"].apply(level)
```

```python
def category(x): 
    if x < 10: 
        return "Light" 
    elif x <= 20: 
        return "Moderate" 
    else: 
        return "Heavy" 

df["Category"] = df["weekly_self_study_hours"].apply(category)
```
