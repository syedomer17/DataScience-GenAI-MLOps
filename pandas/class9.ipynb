```python
import pandas as pd 
import numpy as np
```

```python
data = {
    "Name":["Alice","Bob","Charlie","David","Emma"],
    "Math":[88,75,92,60,85],
    "Science":[90,80,78,70,95]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df["Total_Marks"] = df[["Math","Science"]].apply(lambda row : row.sum(),axis=True)
```

```python
df
```

```python
col_mean = df[["Math","Science"]].apply(lambda col : col.mean())
```

```python
col_mean
```

```python
def to_gread(Score): 
    if Score >= 90: 
        return "A"
    elif Score >= 80: 
        return "B"
    elif Score >= 70: 
        return "C" 
    elif Score >= 40: 
        return "D"
    else: 
        return "Fail" 

df["Maths Gread"] = df["Math"].apply(to_gread)
```

```python
df
```

```python
df["Name"] = df["Name"].map(str.lower)
```

```python
df
```

```python
df["Bouns Math"] = df["Math"].map(lambda x: x + 5)
```
