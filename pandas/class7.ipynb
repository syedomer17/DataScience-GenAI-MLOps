```python
import pandas as pd 
```

```python
# Dataset 1: Product details
df1 = pd.DataFrame({
    "ProductID": [101, 102, 103],
    "ProductName": ["Pen", "Notebook", "Marker"],
    "Price": [10, 50, 30]
})

# Dataset 2: More products (same columns as df1)
df2 = pd.DataFrame({
    "ProductID": [104, 105],
    "ProductName": ["Pencil", "Eraser"],
    "Price": [5, 8]
})

# Dataset 3: Additional info (different columns)
df3 = pd.DataFrame({
    "Stock": [100, 60, 80],
    "Rating": [4.5, 4.7, 4.8]
})
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df3])
```

```python
pd.concat([df1,df3],axis = True,ignore_index = False)
```

```python
products = pd.concat([df1,df2],ignore_index=False)
```

```python
products
```

```python
sales = pd.DataFrame({
    "ProductID": [101,102,104,106],
    "UnitsSold": [500,200,150,50]
})
```

```python
pd.merge(products,sales, on = "ProductID")
```

```python
pd.merge(products,sales, on = "ProductID", how = "inner")
```

```python
pd.merge(products,sales, on = "ProductID", how = "left")
```

```python
pd.merge(products,sales, on = "ProductID", how = "right")
```

```python
pd.merge(products,sales, on = "ProductID", how = "outer")
```

```python
# Sales data
sales = pd.DataFrame({
    "ProductID": [201, 202, 203, 204],
    "UnitsSold": [500, 300, 400, 250]
})

# Revenue data (missing ProductID 204, extra 205)
revenue = pd.DataFrame({
    "ProductID": [201, 202, 203, 205],
    "Revenue": [7500, 4500, 6800, 1200]
})
```

```python
merged_df = pd.merge(sales,revenue,on="ProductID",how="outer")
merged_df
```

```python
pd.merge(sales,revenue,on="ProductID",how="left")
```

```python
pd.merge(sales,revenue,on="ProductID",how="right")
```

```python
pd.merge(sales,revenue,on="ProductID",how="inner")
```
