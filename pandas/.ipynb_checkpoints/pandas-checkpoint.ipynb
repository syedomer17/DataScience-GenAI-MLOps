```python
import pandas as pd 
import numpy as np
```

```python
s = pd.Series([10,20,30,40])
print(s)
print(s[0])
```

```python
s = pd.Series([100,200,300], index=["a","b","c"])
print(s)
print(s["b"])
```

```python
data = {
    "Math": 90,
    "Science": 85,
    "English": 88
}

s = pd.Series(data)
print(s)
```

```python
data = {
    "Name": ["Omer", "Ali", "Sara"],
    "Age": [20,21,22],
    "City": ["Hyderabad", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)
print(df)
print(df["Name"])
print(df[["Name", "City"]])
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
```

```python
arr = np.array([1,2,3,4])
s = pd.Series(arr)
print(s)
```
