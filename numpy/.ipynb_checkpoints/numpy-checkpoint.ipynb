```python
import numpy as np

sales = np.array([120, 340, 560, 230, 890, 450])
print(sales.sum())
print(sales.mean())
print(np.median(sales))
print(sales.min())
print(sales.max())
print(sales.argmax())
print(sales.argmin())
print(sales.std())
print(sales.var())
print(np.sort(sales))
print(np.sort(sales)[::-1])
```
