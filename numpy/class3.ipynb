```python
import numpy as np 

a = np.array([10,20,30,40])
b = np.array([1,2,3,4])

print(a+b)
print(a*b)
print(a // b)
print(a - b)
```

```python
a = np.array([1,2,3,4])
print(a*2)
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr * 5)
```

```python
arr = np.array([4,8,15,23,42])

print(np.sum(arr))
print(np.mean(arr))
print(np.min(arr))
print(np.max(arr))
print(np.std(arr))
print(np.var(arr))
```
