```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(42)
```

```python
df = pd.DataFrame({
    "Employee": [f"Emp{i}" for i in range(1,21)],
    "Age": np.random.randint(20, 60, 20),
    "Salary": np.random.randint(30000, 100000, 20),
    "Department": np.random.choice(
        ["HR", "IT", "Finance", "Sales"],
        20
    )
})
```

```python
df
```

```python
dept_salary = df.groupby("Department")["Salary"].mean()

dept_salary.plot(
    kind="bar",
    figsize=(8,4),
    title="Average Salary by Department"
)

plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
dept_salary.plot(
    kind="barh",
    figsize=(8,4)
)
plt.title("Department Salary")
plt.show()
```

```python
df["Department"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Department Distribution")
plt.ylabel("")
plt.show()
```

```python
df["Salary"].plot(
    kind="hist",
    bins=8,
    figsize=(8,4)
)

plt.title("Salary is tribution")
plt.show()
```

```python
df.boxplot(column="Salary")
plt.title("Salary Boxplot")
plt.show()
```

```python
df.plot(
    kind="scatter",
    x="Age",
    y="Salary",
    figsize=(8,4)
)
plt.title("Age vs Salary")
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.scatter(df["Age"], df["Salary"])

highest = df["Salary"].idxmax()

plt.annotate(
    df.loc[highest,"Employee"],
    (
        df.loc[highest,"Age"],
        df.loc[highest,"Salary"]
    )
)

plt.show()
```

```python
plt.style.use("ggplot")

df["Salary"].plot(kind="hist")

plt.show()
```

```python
salary_mean = df.groupby("Department")["Salary"].mean()
salary_std = df.groupby("Department")["Salary"].std()

plt.errorbar(
    salary_mean.index,
    salary_mean.values,
    yerr=salary_std.values,
    fmt="o"
)

plt.show()
```
