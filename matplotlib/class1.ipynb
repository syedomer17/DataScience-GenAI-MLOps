```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.xlabel("counter")
plt.ylabel("marks")
plt.title("random")

plt.plot(y,x)
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
plt.figure(figsize=(8,4))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("Age")
plt.ylabel("Score")
plt.grid(True)
plt.title("Random titla")
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(8,4))
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.xlabel("Age")
plt.ylabel("salary")
plt.subplot(2,2,4)
plt.plot(x,y2)
plt.subplot(2,2,2)
plt.plot(y1,y2)
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

fig = plt.figure(figsize=(8,6))
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.5,0.4,0.4])
axis.plot(x,y1)
axis.plot(x,y2)
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.figure(figsize=(8,6),dpi=120)
plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(7,4))

plt.plot(x,y1,color="blue",marker="o",label="Squares",markersize=8)

plt.plot(x,y2,color="red",marker="s",label="Line",markersize=6)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Example: Legends Labels & Title")
plt.legend(loc = "upper left")
```

```python
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]

fig,ax = plt.subplots(figsize=(7,4))
ax.plot(x,y,color="teal",marker="o",linewidth=2)
ax.set_xlim(-1,6)
ax.set_ylim(-5,30)
```

```python
x = [1,2,3,4,5]
y = [5,7,4,6,8]

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x,y)
plt.subplot(1,3,2)
plt.hist(x)
plt.subplot(1,3,3)
plt.boxplot(x)
```
